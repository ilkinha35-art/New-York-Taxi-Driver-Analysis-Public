# NYC Taxi Game 3D - Notebook de Demonstração
Este notebook mostra conceitos, protótipos e ideias relacionadas ao desenvolvimento do jogo.

```python
taxi_position = [0, 0]
def move_taxi(direction):
    if direction == 'up': taxi_position[1] += 1
    elif direction == 'down': taxi_position[1] -= 1
    elif direction == 'left': taxi_position[0] -= 1
    elif direction == 'right': taxi_position[0] += 1
    return taxi_position

print(move_taxi('up'))
print(move_taxi('right'))
```

```python
import matplotlib.pyplot as plt
x, y = taxi_position
plt.scatter(x, y, c='yellow', s=200, label='Taxi')
plt.title('Posição do Taxi (Protótipo)')
plt.legend()
plt.show()
```
